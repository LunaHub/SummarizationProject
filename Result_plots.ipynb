```
# Import libaries:
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Sequence lengths
Seq_len = [0, 10, 15, 20, 25, 30, 35, 40, 45]

# Continuous sequences:
Acc_con = [100, 100, 72, 51, 39, 26, 11, 6.9, 2.7]
Loss_con = [0, 0.05, 0.06, 0.2, 0.4, 0.75, 1.6, 1.6, 3.4]

# Random order:
Acc = [100, 95.5, 58.5, 34.4, 24.5, 19.1, 13.9, 10.4, 8.1]
Loss = [0, 0.17, 1.6, 14.9, 22.8, 15.2, 26.5, 91, 80]

# Dumme data: (not done testing...)
Acc_attn = [100, 100, 98.7, 95.6, 90, 91.99, 72.2, 76.5, 49.6]
Loss_attn = [0, 0.0001, 0.0115, 0.0237, 1.3, 0.2, 1.9, 2.1,3.4]
```

```
# plt.figure(figsize = [8,4])
plt.rcParams.update({'font.size': 16})

# Continuous seq vs random seq:
plt.plot(Seq_len, Acc_con, label = 'Continuous sequence')
plt.plot(Seq_len, Acc, label = 'Random sequence')
plt.legend()
plt.xlabel('Maximum sequence length')
plt.ylabel('[%]')
plt.title('Validation accuracy');
plt.tight_layout()
plt.savefig('Accuracy_con_vs_rand.png', dpi = 500)
plt.show()

plt.plot(Seq_len, Loss_con, label = 'Continuous sequence')
plt.plot(Seq_len, Loss, label = 'Random sequence')
plt.legend()
plt.xlabel('Maximum sequence length')
plt.ylabel('Loss')
plt.title('Training loss');
plt.tight_layout()
plt.savefig('Loss_con_vs_rand.png', dpi = 500)
plt.show()

# Attention vs no attention

plt.plot(Seq_len, Acc, label = 'No Attention')
plt.plot(Seq_len, Acc_attn, label = 'Attention')
plt.legend(loc = 'lower left')
plt.xlabel('Maximum sequence length')
plt.ylabel('[%]')
plt.title('Validation accuracy');
plt.tight_layout()
plt.savefig('Accuracy.png', dpi = 500)
plt.show()

plt.plot(Seq_len, Loss, label = 'No Attention')
plt.plot(Seq_len, Loss_attn, label = 'Attention')
plt.legend()
plt.xlabel('Maximum sequence length')
plt.ylabel('Loss')
plt.title('Training loss');
plt.tight_layout()
plt.savefig('Loss.png', dpi = 500)
plt.show()
```
